```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
%matplotlib inline
```

```python
x=np.linspace(1,10,10)
y=x**2
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.title('First Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
```

```python
plt.subplot(4,1,4)
plt.plot(x,y, 'red')
plt.subplot(4,2,4)
plt.plot(x,y,'green')
plt.show()
```

```python
plt.subplot()
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,'purple')
ax1=fig.add_axes([0,0,0.6,0.6])
ax2=fig.add_axes([0.9,0.9,0.9,0.9])
ax1.plot(x,y,'red')
ax2.plot(x,y,'green')
```

```python
fig,ax=plt.subplots(nrows=2,ncols=2)
plt.tight_layout()
```

```python
x=np.array([1,2,3,4,5,6,7])
y=x**2
fig,ax=plt.subplots(nrows=2,ncols=2)
plt.tight_layout()
ax[0,1].plot(x,y,'orange')
ax[1,0].plot(x,y,'grey')
ax[0][1].set_xlabel('x label 1')
ax[1][0].set_xlabel('x label 2 ')
ax[0][1].set_ylabel('y label 1')
ax[1][0].set_ylabel('y label 2 ')
ax[0][1].set_title('Axis [0,1]')
ax[1][0].set_title('Axis [1,0] ')
```

```python
fig,ax=plt.subplots(nrows=2,ncols=1)
plt.tight_layout(pad=3.0)
ax[0].plot(x,y,'orange')
ax[1].plot(x,y,'grey')
ax[0].set_xlabel('x label 1')
ax[1].set_xlabel('x label 2 ')
ax[0].set_ylabel('y label 1')
ax[1].set_ylabel('y label 2 ')
ax[0].set_title('Axis [0]')
ax[1].set_title('Axis [1] ')
```

## Stacked Bar Chart

```python
men=[5,6,7,8]
women=[1,2,3,4]
n=np.arange(4)
plt.bar(n,men,color='purple')
plt.bar(n,women,color='pink')
plt.legend(['men','women'])
```

## Bar Chart

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(n,men,color='grey',width=0.5)
ax.bar(n+0.5,women,color='green', width=0.5)
ax.legend(['men','women'])
#ax.text(10,10,'str')
#ax.annotate('annotation',(1,2))
```

## Pie Chart

```python
label = ['Dog','Cat','Wolf','Lion']
sizes = [50,45,60,80]

#creating the pie chart 
plt.pie(sizes, labels = label,autopct='%1.1f%%',shadow=True, startangle = 90,explode = (0,0.1,0,0))
plt.title("Pie Chart Demo")
plt.show()
```

## Box Plot

```python
data=np.random.random(50)
plt.boxplot(data,showmeans=True)
plt.grid(True)
plt.show()
```

## Violin Plot

```python
data=np.random.random(25)
plt.violinplot(data,showmeans=True,showmedians=True)
plt.show()
```

## Area Plot

```python
#define the dataset
x=range(1,15)
y=[1,4,6,8,4,5,3,2,4,1,5,6,8,7]

#creating the area plot
plt.stackplot(x,y,colors = 'Black', alpha = 0.5)
#Merge area chart with the line chart
plt.plot(x,y,color = 'orange')
plt.grid(True)
plt.show()
```

## Histogram

```python
#numbers=np.random.random(50)
numbers = [10,12,16,19,11,20,26,28,30,38,35,34,45,60,68,64,62,70,78,75,79,85,94,95]
# creating the histogram
plt.hist(numbers, bins = [0,20,40,60,80,100], edgecolor='#000000', color ='green')
plt.title("Histogram Demo")
plt.xlabel("Range of values")
plt.ylabel("Frequence of values")
plt.grid(True)
plt.savefig("Histogram Plot.png")
plt.show()
```
