## Stacked Bar Chart

```python
men=[4,5,6,7,8]
women=[1,2,3,4,5]
n=np.arange(5)
plt.bar(n,men,color='brown')
plt.bar(n,women,color='pink')
plt.show()
```

```python
plt.bar(n,men,color='brown',width=0.5)
plt.bar(n+0.5,women,color='pink',width=0.5)
plt.xlabel('Male and Female')
plt.ylabel('Count')
plt.legend(['men','women'])
plt.show()
```

## Pie Chart

```python
label=['A+','A','B+','B','C','D']
numbers=[20,34,54,27,45,65]
plt.pie(numbers,labels=label)
plt.show()
```

```python
plt.pie(numbers,labels=label,explode=(0,0.2,0,0,0.2,0),shadow=True)
plt.show()
```

## Sub Plots

```python
plt.subplot(2,1,1)
x=np.linspace(1,10,10)
y=x**2
plt.plot(x,y,color='red')
plt.subplot(2,1,2)
plt.plot(x,y,color='green')
plt.show()
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,'purple')
ax1=fig.add_axes([1.5,1.5,0.5,0.5])
ax1.plot(x,y,'orange')
plt.show()
```

```python
fig,ax=plt.subplots(nrows=2,ncols=2)
plt.tight_layout(pad=4.5)
ax[0,1].plot(x,y,'lightblue')
ax[1,0].plot(x,y,'brown')
ax[0][1].set_xlabel('X lable [0,1]')
ax[1][1].set_title('Title of [1,1]')
plt.show()
```
